```python
# Load libraries
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt

# Select nbagg to use as matplotlib backend
matplotlib.use('nbagg')
```

```python
# Create DataFrame containing the trivia point data
trivia_df = pd.read_csv('all-points-anonymized.csv')
trivia_df = trivia_df.set_index('Date')

# Printing to gain basic information of data from CSV
print('Number of columns:', trivia_df.shape[1])
print('Number of rows:', trivia_df.shape[0])
trivia_df.head()
```

```python
# Print the dtypes to find columns that need type conversions
print(trivia_df.dtypes)
```

```python
# Function to convert percentage objects to floats
def percent_object_to_float(x):
    # Remove '%' character, convert object to float, divide by 100
    return float(x.strip('%'))/100

# Convert '% Right' column to floats
trivia_df['% Right'] = trivia_df['% Right'].apply(percent_object_to_float)
# Convert '% Points' column to floats
trivia_df['% Points'] = trivia_df['% Points'].apply(percent_object_to_float)
```
